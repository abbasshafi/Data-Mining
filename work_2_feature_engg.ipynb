### Feature Engineering

```python
df.shape
```

**Total Activity**

```python
df['TotalActivity'] = df['Administrative'] + df['Informational'] + df['ProductRelated']
df['TotalActivity'][:5]
```

```python
df.shape
```

**Total Duration**

```python
df['TotalDuration'] = (
    df['Administrative_Duration'] +
    df['Informational_Duration'] +
    df['ProductRelated_Duration']
)
df['TotalDuration'][:5]
```

**Avg Time Per Page**

```python
df['AvgTimePerPage'] = df['TotalDuration'] / (df['TotalActivity'] + 1)
df['AvgTimePerPage'][:3]
```

**Engagement Score**

```python
df['EngagementScore'] = df['PageValues'] * (1 - df['BounceRates'])
```

**Exit-Bounce Difference**

```python
df['ExitBounceDiff'] = df['ExitRates'] - df['BounceRates']
```

**Product Focus Ratio**

```python
df['ProductFocus'] = df['ProductRelated'] / (df['TotalActivity'] + 1)
```

```python
df.shape
```

```python
df.drop([
    'Administrative', 
    'Informational', 
    'ProductRelated',
    'Administrative_Duration', 
    'Informational_Duration', 
    'ProductRelated_Duration'
], axis=1, inplace=True)
```

**One hot encode**

```python
categorical_cols = [
    'Month', 'VisitorType', 'OperatingSystems',
    'Browser', 'Region', 'TrafficType'
]
for i in categorical_cols:
    print(i,df[i].nunique())
```

```python
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
df.head(3)
```

## Split Data to X,Y

```python
X = df.drop('Revenue', axis=1)
X.head(3)
```

```python
y = df['Revenue']
y.head(3)
```

## Standardization

*standardization is to fix the scale for all features*

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

## Train Test Split

```python
print((y==1).sum())
print((y==0).sum())
```

*as we know the target column i.e revenue is imbalanced so we will do split through stratified TTS (hence each split have True and false then )*

*otherwise its have chances during the split that one class have no or less number of (True's)*

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    stratify=y,
    random_state=1
)
```

```python
print((y_train==0).sum())
print((y_train==1).sum())
print((y_test==0).sum())
print((y_test==1).sum())
```

### Logistic Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(class_weight='balanced', max_iter=1000)
```

```python
model.fit(X_train, y_train)
```

### Model Saving

```python
import joblib

## model saving
joblib.dump(model, "logistic_model.pkl")

## saving scaled data as i train model on scaled data 
joblib.dump(scaler, "scaler.pkl")
```

### Model Evaluation for Logistic Regression

```python
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

```python
y_pred = model.predict(X_test)
y_pred[:3]
```

_____________

**Accuracy**

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

__________

**Error_rate**

```python
error_rate = 1-accuracy
print("Error Rate: ",error_rate)
```

_____________

**Precision**

```python
precision = precision_score(y_test, y_pred)
print("Precision:", precision)
```

____________

**Recall**

```python
from sklearn.metrics import recall_score
```

```python
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
```

**F-1**

```python
from sklearn.metrics import f1_score
```

```python
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```

-----------

```python
print("Confusion Matrix for Logistic Regression:\n", confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```

**ROC-AUC & PR-AUC for LR**

```python
from sklearn.metrics import roc_auc_score, average_precision_score
```

```python
y_prob_lr = model.predict_proba(X_test)[:, 1]
```

```python
roc_auc_lr = roc_auc_score(y_test, y_prob_lr)
pr_auc_lr = average_precision_score(y_test, y_prob_lr)

print("Logistic Regression ROC-AUC:", roc_auc_lr)
print("Logistic Regression PR-AUC:", pr_auc_lr)
```

### Random Forest Model Training
